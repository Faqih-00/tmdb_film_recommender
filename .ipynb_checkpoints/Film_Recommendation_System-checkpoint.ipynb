## Import our dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import word_tokenize
from sklearn.metrics.pairwise import cosine_distances
```

## Import dataset

```python
movies = pd.read_csv('data/tmdb_5000_movies.csv')
credits = pd.read_csv('data/tmdb_5000_credits.csv')
movies.head()
```

```python
credits.head()
```

### Merge those dataframe

```python
df = movies.merge(credits, on='title')
df.head()
```

```python
df = df[["movie_id", "title", "overview", "genres", "keywords", "cast", "crew"]]
df.head()
```

#### Missing values

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

#### Preprocessing our feature

```python
df.iloc[0].genres
```

```python
import ast
```

```python
def preprocess(obj):
    result = []
    for i in ast.literal_eval(obj):
        result.append(i['name'])
    return result
```

```python
df['genres'] = df['genres'].apply(preprocess)
df.head()
```

```python
df['keywords'] = df['keywords'].apply(preprocess)
df.head()
```

```python
# preprocess cast column  
def preprocess_cast(obj):
    result = []
    count = 0
    for i in ast.literal_eval(obj):
        if count < 3:
            result.append(i['name'])
        count += 1
    return result
```

```python
df['cast'] = df['cast'].apply(preprocess_cast)
df.head()
```

```python
# take director from crew column
def director(obj):
    result = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            result.append(i['name'])
    return result
```

```python
df['crew'] = df['crew'].apply(director)
df.head()
```

```python
df['overview'] = df['overview'].apply(lambda x : x.split())
df.head()
```

```python
# collapse whitespace from our words
def collapse_ws(l):
    result = []
    for i in l:
        result.append(i.replace(" ", ""))
    return result
```

```python
df['cast'] = df.cast.apply(collapse_ws)
df['crew'] = df.crew.apply(collapse_ws)
df['genres'] = df.genres.apply(collapse_ws)
df['keywords'] = df.keywords.apply(collapse_ws)
df.head()
```

```python
# metadata
df['metadata'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
df.head()
```

```python
df.loc[2972]
```

```python
new_df = df[['movie_id', 'title', 'metadata']]
new_df.head()
```

```python
new_df['metadata'] = new_df['metadata'].apply(lambda x : " ".join(x))
new_df.head()
```

```python
class FilmRecommenderSystem:
    def __init__(self,data, content_col):
        self.df = pd.read_csv(data)
        self.content_col = content_col
        self.encoder = None
        self.bank = None
        
    def fit(self):
        self.encoder = CountVectorizer(stop_words='english', tokenizer=word_tokenize)
        self.bank = self.encoder.fit_transform(self.df[self.content_col])
    
    def recommend(self, idx, top=10):
        content = new_df.loc[idx, self.content_col]
        code = self.encoder.transform([content])
        dist = cosine_distances(code, self.bank)
        rec_idx = dist.argsort()[0, 1:(top+1)]
        return new_df.loc[rec_idx]
```

```python
recsys = FilmRecommenderSystem('tmdb_preprocessed_movies.csv', content_col="metadata")
recsys.fit()
```

```python
recsys.recommend(0)
```

```python
recsys.recommend(1)
```

```python
recsys.recommend(3) 
```
